```python
clarity = pd.get_dummies(auto['clarity'], drop_first = False)
clarity
```

```python
auto.drop(['cut','color','clarity'], axis = 1, inplace = True)
```

```python
# Starting machine learning
import seaborn as sns
plt.figure(figsize=(35,10))
sns.heatmap(auto.corr(),annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
df_train, df_test = train_test_split(auto, train_size = 0.7, test_size = 0.3, random_state = 1)
```

```python
df_train = df_train.iloc[:,1:]
```

```python
df_test = df_test.iloc[:,1:]
```

```python
list(auto.keys())
```

```python
from sklearn.linear_model import LinearRegression
X_train = df_train[['carat',
 'depth',
 'table',
 'x',
 'y',
 'z',
 'Good',
 'Ideal',
 'Premium',
 'Very Good',
 'D',
 'E',
 'F',
 'G',
 'H',
 'I',
 'J',
 'I1',
 'IF',
 'SI1',
 'SI2',
 'VS1',
 'VS2',
 'VVS1',
 'VVS2']]
y_train = df_train[['price']]
# # df_test_predict = X_train.iloc[0:1]
```

```python
X_train
```

```python
y_train
```

```python
lm = LinearRegression()
model = lm.fit(X_train, y_train)
```

```python
# Testing
data = df_test.iloc[0:1]
actual_price = data.pop('price')

print(data)
predicted_price = model.predict(data)
print("Actual price",actual_price)
print("predicted expenses",predicted_price)
print(df_test.iloc[0:1])
```
