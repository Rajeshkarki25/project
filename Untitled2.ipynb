```python
df = df.dropna()
```

```python
df.keys()
```

```python
auto = df
```

```python
CompanyName = pd.get_dummies(df['Company Name'], drop_first = True)
CompanyName
```

```python
auto = pd.concat([auto,CompanyName], axis = 1)
auto
```

```python
job_typ = pd.get_dummies(auto['Job Title'], drop_first = False)
job_typ
```

```python
auto = pd.concat([auto, job_typ], axis = 1)
auto
```

```python
auto
```

```python
auto.drop(['Company Name','Job Title','Location'], axis = 1, inplace = True)
```
